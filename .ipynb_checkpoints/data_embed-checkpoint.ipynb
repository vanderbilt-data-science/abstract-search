```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import numpy as np
```

```python
# Load data from csv
df = pd.read_csv("sessions.csv", usecols=['Unique ID', 'Name', 'Description', 'Activity Code', 'Start Time', 'End Time', 'Location Name'])
df.head()
```

```python
# Use the Description column as the documents to embed
docs = list(df['Description'])

# Load model
model = SentenceTransformer('sentence-transformers/multi-qa-MiniLM-L6-cos-v1')

# Embed documents
doc_emb = model.encode(docs)
```

```python
#Save embeddings to file
np.savetxt("abstract-embed.txt", doc_emb)
```
